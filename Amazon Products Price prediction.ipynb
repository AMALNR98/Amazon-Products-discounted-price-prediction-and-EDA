Replacing String values

```python
amazon['actual_price'] = amazon['actual_price'].str.replace('₹','')
amazon['discounted_price'] = amazon['discounted_price'].str.replace('₹','')
amazon['discount_percentage'] = amazon['discount_percentage'].str.replace('%','').astype('float64')

amazon.head()
```

Changing datatype object to float64 and removing , from prices and rating count

Creating new category column rating_score for better understanding of rating

```python
rating_score = []
for i in features['rating']:
    if i < 2.0: rating_score.append('Very unsatified')
    elif i < 3.0: rating_score.append('Unsatified')
    elif i < 4.0: rating_score.append('Neutral')
    elif i < 5.0: rating_score.append('Satified')
    elif i == 5.0: rating_score.append('Very satified')
```

```python
features['rating_score'] = rating_score
features['rating_score'] = features['rating_score'].astype('category')
# Reorder cateories
features['rating_score'] = features['rating_score'].cat.reorder_categories(['Unsatified', 'Neutral', 'Satified','Very satified'], ordered=True)
features.head()
```

```python
rating_score_pct = features['rating_score'].value_counts(normalize=True).rename_axis('score').reset_index(name='score_pct')
rating_score_pct['score_pct'] = rating_score_pct['score_pct'].round(3)
fig, ax = plt.subplots(1,2, figsize = (15,7))
sns.countplot(ax=ax[0], data=features, x='rating_score')
sns.barplot(ax=ax[1],data = rating_score_pct, x='score', y='score_pct')

ax[0].set_title('Amount of products by Rating Score', fontweight = 'bold')
ax[1].set_title('Percentage of product by Rating Score', fontweight = 'bold')

ax[0].set_xlabel('Rating Score', fontweight = 'bold')
ax[1].set_xlabel('Rating Score', fontweight = 'bold')

ax[0].set_ylabel('Amount', fontweight = 'bold')
ax[1].set_ylabel('Percentage', fontweight = 'bold')

ax[0].bar_label(ax[0].containers[0])
ax[1].bar_label(ax[1].containers[0])
plt.show()
```

Filling null values

```python
features['rating_count'].fillna(features['rating_count'].mode()[0], inplace=True)
```

```python
features.columns
```

```python
x = features[['rating', 'rating_count', 'actual_price','discount_percentage']]
y = features['discounted_price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 42)
```

Ploting scaterplot  and prediction line

```python
import seaborn as sns
sns.regplot(x = features['actual_price'], y = features['discounted_price'], color = 'red')
```

Model creation

```python
from sklearn.linear_model import LinearRegression
# Create Object(model)
model = LinearRegression()
# fit model
model.fit(x_train, y_train)
# Predict value
y_pred = model.predict(x_test)
y_pred
```

```python
print('slope or coefficient is', model.coef_)  # coef_
print("constant is ", model.intercept_)
```

```python
df = pd.DataFrame({'Actual_value':y_test, 'Predicted_value':y_pred, 'Difference': y_test - y_pred})
df
```

```python
# Mean Absoulte Percentage error
from sklearn.metrics import mean_absolute_percentage_error
print("Error percentage is ", mean_absolute_percentage_error(y_test, y_pred))
```

```python
# R2_Score
from sklearn.metrics import r2_score
print("R2 score is", r2_score(y_test, y_pred))
```
